```python
import os
import pickle
import xml.etree.ElementTree as elemTree
import re
from tqdm import tqdm
```

```python
base_dir = './crawl_data/notice_of_feedback/RejectDecision'
```

```python
candidate_pkl = os.listdir(base_dir)
len(candidate_pkl)
```

```python
count_pkl = {}
use_pkl = []
for _pkl in tqdm(candidate_pkl):
    load_pkl_dir = os.path.join(base_dir, _pkl)
    _use_pkl = {}
    with open(load_pkl_dir, 'rb') as f:
        data = pickle.load(f) 
    _data = 'fail' if data.count('MAXIMUM 1000 COUNT OVER') == 1 else data
    if count_pkl.get(_data, None) == None:
        count_pkl[_data] = 1
    else:
        count_pkl[_data] += 1
    if _data != 'fail':
        _use_pkl[_pkl] = _data
        use_pkl.append(_use_pkl)
```

```python
ns = {"soapenv" : "http://schemas.xmlsoap.org/soap/envelope/",
      "ax264":"http://intermediateDocument.bean.services.webservice.plus.kipris.or.kr/xsd",
      "ns" : "http://plus.kipris.or.kr/xsd"}
```

```python
available = []
query_id = []
file_id = []

server_fail = 'The server is currently unable to handle the request due to a temporary overloading or maintenance of the server.'
fail_id = []

for candidate in tqdm(case_extract_documents):
    file_name = list(candidate.keys())[0]
    file_content = list(candidate.values())[0]
    candidate = re.sub('&#[0-9]+;', '', file_content)
    if candidate.count(server_fail) > 0:
        fail_id.append({'file_name': file_name, 'candidate':candidate})
        continue
    tree = elemTree.fromstring(candidate)
    content = tree.findall("./soapenv:Body/ns:rejectDecisionResponse/ns:return/ax264:rejectionDetail1", ns)
    id_number = tree.findall("./soapenv:Body/ns:rejectDecisionResponse/ns:return/ax264:applicationNumber", ns)
    if len(content) > 0:
        for _c in content:
            available.append(_c.text)
            query_id.append(id_number[0].text)
            file_id.append(file_name)
```

```python
result = []
candidate_pattern = '4{1}[0-9-]{8,2000}'

for q, a, f in tqdm(zip(query_id, available, file_id)):
    if a == None:
        continue
    for _a in a.split('.'):
        group_candidate = re.findall(candidate_pattern, _a)
        if len(group_candidate) > 0:
            _result = {}
            _result['query'] = q
            _result['reason'] = '칭호,관념,외관'
            _result['file_name'] = f
            group_candidate = [x for x in group_candidate if x != f]
            for _group_candidate in group_candidate:
                _result['candidate'] = _group_candidate
                result.append(_result)
        else:
            continue
```

```python
query_set = [i['query'] for i in result]
candidate_set = [i['candidate'] for i in result]
```
